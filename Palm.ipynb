```python
def convert_date(date_str):
    thai_months = {
        "ม.ค.": "01",
        "ก.พ.": "02",
        "มี.ค.": "03",
        "เม.ย.": "04",
        "พ.ค.": "05",
        "มิ.ย.": "06",
        "ก.ค.": "07",
        "ส.ค.": "08",
        "ก.ย.": "09",
        "ต.ค.": "10",
        "พ.ย.": "11",
        "ธ.ค.": "12"
    }
    day, month_abbr, year = date_str.split('-')
    month = thai_months[month_abbr]
    return f"{'0' if len(day) == 1 else ''}{day}-{month}-{year[-2:]}"

# ทดสอบการใช้งานฟังก์ชัน
print(convert_date("21-ธ.ค.-18"))  # ผลลัพธ์: 21-12-18
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your data is already loaded into a DataFrame called selectPL

# Convert PRICE_DATE column to datetime if it's not already in datetime format
selectPL['PRICE_DATE'] = pd.to_datetime(selectPL['PRICE_DATE'], format='%d-%m-%y')

# Group data by month and calculate average price for each month
average_price_per_month = selectPL.groupby(selectPL['PRICE_DATE'].dt.to_period('M'))['PRICE_DAY'].mean()

# Plot the average price for each month
plt.figure(figsize=(10, 6))
plt.plot(average_price_per_month.index.astype(str), average_price_per_month, marker='o', linestyle='-')
plt.xlabel('Month')
plt.ylabel('Average Price (THB)')
plt.title('Average Price Per Month')

# Annotate each data point with its corresponding value
for i, price in enumerate(average_price_per_month):
    plt.text(average_price_per_month.index[i].strftime('%Y-%m'), price, str(round(price, 2)), ha='right')

plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
